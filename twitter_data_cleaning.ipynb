### clean_func()

```python
punc = string.punctuation
sw = stopwords.words('english')
sw+=punc
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    if tag.startswith('V'):
        return wordnet.VERB
    if tag.startswith('N'):
        return wordnet.NOUN
    if tag.startswith('R'):
        return wordnet.ADV
    return wordnet.NOUN
```

```python
# lemmatizing
lem = WordNetLemmatizer()

def clean_review(docs):
    output=[]
    docs = re.sub("[^a-zA-Z]", " ", docs)
    print(docs)
    for word in word_tokenize(docs):
        if word.lower() not in sw and len(word.lower())>3:
            clean_word = lem.lemmatize(word,pos = get_simple_pos(pos_tag(word)[0][1]))
            output.append(clean_word)
    return output
```

```python
data=[clean_review(doc) for doc in data]
```

```python
data
for i in range(len(data)):
    data[i]=(" ".join(data[i])).lower()
```
